## Importing necessary libraries

```python
from selenium import webdriver

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from bs4 import BeautifulSoup
import requests

import time
import pandas as pd
```

## Scraping the data

```python
PATH = "C:\\Program Files (x86)\\chromedriver.exe"
driver = webdriver.Chrome(PATH)

driver.get("https://www.flipkart.com/")

try:
    # Closing the login popup-
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "body > div._2Sn47c > div > div > button"))
    )
    element.click()
    
    # Clicking on Mobiles Section-
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#container > div > div._331-kn._2tvxW > div > div > div:nth-child(3) > a > div._1mkliO > div > img"))
    )
    element.click()
    
    # Clicking on Mobiles to get the list of all the mobiles-
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#container > div > div._6t1WkM._3HqJxg > div._1YokD2._2GoDe3 > div._1AtVbE.col-2-12 > div > div > section > div:nth-child(3) > div > a"))
    )
    element.click()
    
    # Creating empty lists to store the scraped data-
    names=[]
    price=[]
    info=[]
    
    # Defining the number of pages to be scraped-
    pages_to_visit = 3
    
    # Looping through all the pages-
    for i in range(pages_to_visit):   
        # To scrape the data, we need to pass the URL
        url = driver.current_url
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        
        # Scraping the Model names-
        find = soup.find_all("div",class_= "_4rR01T")
        for i in find:
            names.append(i.get_text())
        
        # Scraping the prices-
        find = soup.find_all("div",class_= "_30jeq3 _1_WHN1")
        for i in find:
            price.append(i.get_text()[1:])
        
        # Scraping the Specifications-
        find = soup.find_all("ul",class_= "_1xgFaf")
        for i in find:
            info.append(i.get_text())
        
        # Clicking on the next page-
        element = WebDriverWait(driver, 10).until(     # To make sure the elements are loaded, we'll wait for max. 10 seconds
            EC.presence_of_element_located((By.LINK_TEXT, "NEXT"))
        )
        element.click()
        
except:
    print("Something went Wrong!")
    
finally:
    driver.quit()
```

## Appending our scraped data in the dataframe

```python
df = pd.DataFrame({'Model Name':names, 'Price (in Rs.)':price, 'Specs':info})
```
